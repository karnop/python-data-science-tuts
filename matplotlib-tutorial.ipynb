```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # we'll just use its dataset
```

# 2D Line plots
- used in bivariate analysis (when we have 2 cols)
- categorical-numerical or numerical-numerical
- use cases : TimeSeries analysis

```python
# plotting 
price = [48000, 54000, 57000, 49000, 47000, 45000]
year = [2015, 2016, 2017, 2018, 2019, 2020]

plt.plot(year, price)
```

```python
# from a pandas dataframe
batsman = pd.read_csv("/kaggle/input/dataset1/sharma-kohli.csv")
batsman
```

```python
plt.plot(batsman["index"], batsman["V Kohli"])
```

```python
# plotting multiple plots
plt.plot(batsman["index"], batsman["V Kohli"])
plt.plot(batsman["index"], batsman["RG Sharma"])
```

```python
# labelling
plt.plot(batsman["index"], batsman["V Kohli"])
plt.plot(batsman["index"], batsman["RG Sharma"])
plt.title("Rohit Sharma vs Virat Kohli runs")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
# colors
plt.plot(batsman["index"], batsman["V Kohli"], color="red") # we can also give hex code for the color
plt.plot(batsman["index"], batsman["RG Sharma"], color="blue")
plt.title("Rohit Sharma vs Virat Kohli runs")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
# line width and style
plt.plot(batsman["index"], batsman["V Kohli"], color="red", linestyle="dashed") # default is solid
plt.plot(batsman["index"], batsman["RG Sharma"], color="blue", linestyle="dotted", linewidth=3)
plt.title("Rohit Sharma vs Virat Kohli runs")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
# markers
plt.plot(batsman["index"], batsman["V Kohli"], color="red", marker="D")
plt.plot(batsman["index"], batsman["RG Sharma"], color="blue", marker="o")
plt.title("Rohit Sharma vs Virat Kohli runs")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
# we also have the marker size property
```

```python
# legend 
plt.plot(batsman["index"], batsman["V Kohli"], color="red", marker="D", label="Virat")
plt.plot(batsman["index"], batsman["RG Sharma"], color="blue", marker="o", label="Rohit")
plt.title("Rohit Sharma vs Virat Kohli runs")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
plt.legend() # this shows the labels
```

```python
# limiting axes 
price = [48000, 54000, 57000, 49000, 47000, 45000, 4500000]
year = [2015, 2016, 2017, 2018, 2019, 2020, 2021]

plt.plot(year, price, color="violet") # this graph doesnot shows variation so we limit the y axis
plt.ylim(0, 100000)
plt.plot(year, price)
```

```python
# grid
plt.plot(batsman["index"], batsman["V Kohli"], color="red", marker="D", label="Virat")
plt.plot(batsman["index"], batsman["RG Sharma"], color="blue", marker="o", label="Rohit")
plt.title("Rohit Sharma vs Virat Kohli runs")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
plt.legend()
plt.grid()
```

```python
# show : we have to call this command to plot graphs in code editors or websites
plt.plot(batsman["index"], batsman["V Kohli"], color="red", marker="D", label="Virat")
plt.plot(batsman["index"], batsman["RG Sharma"], color="blue", marker="o", label="Rohit")
plt.title("Rohit Sharma vs Virat Kohli runs")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
plt.legend()
plt.grid()
plt.show()
```

# Scatter plots
- Bivariate analysis (2 cols)
- numerical vs numerical
- Use case - Finding correlation

```python
# plotting
x = np.linspace(-10, 10, 50)
y = 10*x + 3 + np.random.randint(0, 300, 50)
plt.scatter(x,y)
```

```python
# using a pandas dataframe
batter = pd.read_csv("/kaggle/input/dataset1/batter.csv")
batter = batter.head(50) # top 50 batsmen
```

```python
plt.scatter(batter["avg"], batter["strike_rate"])
plt.title("Avg and SR analysis of top 50 batsmen")
plt.xlabel("Average")
plt.ylabel("Strike Rate")
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
# size : weightage of each point
plt.scatter(tips["total_bill"], tips["tip"], s=tips["size"]*20)
plt.xlabel("Total Bill")
plt.ylabel("Tip Amount")
```

```python
# plotting using plt.plot
plt.plot(tips["total_bill"], tips["tip"], "o") # the o converts 2d plot to scatter plot
```

# Bar Chart 
- Bivariate or univariate analysis
-  Numerical vs categorical
- Use Case: Aggregate analysis of groups

```python
# simple bar chart
children = [10, 20, 40, 10, 30]
colors = ["Red", "Blue", "Green", "Yellow", "Pink"]
plt.bar(colors, children)
```

```python
# Horizontal bar chart
plt.barh(colors, children)
```

```python
records = pd.read_csv("/kaggle/input/dataset1-5/batsman_season_record.csv")
records
```

```python
# multiple bar charts
plt.bar(np.arange(records.shape[0]) -0.3, records["2015"], width=0.3)
plt.bar(np.arange(records.shape[0]), records["2016"], width=0.3)
plt.bar(np.arange(records.shape[0]) + 0.3, records["2017"], width=0.3)

# xticks : replaces stuff on x axis
plt.xticks(np.arange(records.shape[0]), records["batsman"]) 
```

```python
# rotation : helps in label overlapping
plt.bar(np.arange(records.shape[0]) -0.3, records["2015"], width=0.3)
plt.bar(np.arange(records.shape[0]), records["2016"], width=0.3)
plt.bar(np.arange(records.shape[0]) + 0.3, records["2017"], width=0.3)
plt.xticks(np.arange(records.shape[0]), records["batsman"], rotation="vertical") 
```

```python
# stacked bar chart
plt.bar(records["batsman"], records["2015"], bottom=(records["2016"]+records["2017"]), color="red", label="2015")
plt.bar(records["batsman"], records["2016"], bottom=records["2017"], color="blue", label="2016")
plt.bar(records["batsman"], records["2017"], color="green", label="2017")
plt.legend()
```

# Pie chart
- univariate/bivariate analysis
- Categorical vs Numerical

```python
# simple data
data = [23,45,100,20,49]
plt.pie(data)
```

```python
data = [23,45,100,20,49]
subjects = ["Hindi", "SST", "Mathematics", "Science", "Social Work"]
plt.pie(data, labels=subjects)
plt.show() # removes extra info at top
```

```python
# dataset
df = pd.read_csv("/kaggle/input/dataset1/gayle-175.csv")
df
```

```python
# percentage
plt.pie(df["batsman_runs"], labels=df["batsman"], autopct="%0.1f%%")
plt.show()
```

```python
# explode
plt.pie(df["batsman_runs"], labels=df["batsman"], autopct="%0.1f%%", explode=[0.1, 0,0,0,0,0])
plt.show()
```
