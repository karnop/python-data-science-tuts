# Numpy tutorial walkthrough with DSMP

```python
# creating a numpy array
import numpy as np

a = np.array([1,2,3])
print(a)
print(type(a))
```

```python
#2d and 3d array

b = np.array([[1, 2, 3],[4,5,7]])
print(b, "\n")

c = np.array([[[1,2],[3,4]], [[5,6],[7,8]]])
print(c)
```

```python
#dtype : set a datatype for the array
a = np.array([0, 1, 3, 5], dtype=float)
b = np.array([0, 1, 3, 5], dtype=bool)
c = np.array([0, 1, 3, 5], dtype=complex)
print(a)
print(b)
print(c)
```

```python
# np.a-range
a = np.arange(1, 10, )
print(a)
a = np.arange(1, 10, 2)
print(a)


#arange with reshape
b = np.arange(1, 11).reshape(5, 2)
print(b)

#making 4d array
c = np.arange(16).reshape(2,2,2,2)
print("c : \n",c)
```

```python
#np.ones and #np.zeroes and #np.random
ones = np.ones((3, 4))
print(ones)

zeros = np.zeros((3, 4))
print(zeros)

rand = np.random.random((3, 4))
print(rand)
```

```python
#np.linspace : linear space : random numbers between a range, at equal distance
a = np.linspace(-10, 10, 10)
print(a)
```

```python
#np.identity
a = np.identity(3)
print(a)
```

## Numpy array function

```python
a = np.random.random((3,3))
a = np.round(a*100)
print(a)
```

```python
# max, min, sum, prod
print(np.max(a))
print(np.min(a))
print(np.sum(a))
print(np.prod(a))

#axis : 0 = column, 1 = row
print(np.max(a, axis = 1))
print(np.max(a, axis = 0))
```

```python
# mean, median, std, variance
print(np.mean(a))
print(np.median(a))
print(np.std(a))
print(np.var(a))
```

```python
# trigonometric functions
print(np.sin(a), "\n")
print(np.cos(a), "\n")
print(np.tan(a), "\n")
```

```python
# dot product
b = np.arange(12).reshape(3,4)
c = np.arange(12, 24).reshape(4,3)

print(np.dot(b, c))
```

```python
#log and exponents
print(np.log(a))
print(np.exp(a))
```

```python
# round, floor, ceil
c = np.random.random((2,3))
print(c)
d = np.round(c*100)
print(d)
d = np.floor(c*100)
print(d)
d = np.ceil(c*100)
print(d)
```

## Indexing and Slicing

```python
a = np.arange(10)
b = np.arange(12, dtype=float).reshape(3,4)
c = np.arange(27).reshape(3,3,3)
```

```python
print(a)

#negative indexing
print(a[-1])
```

```python
print(b)

# getting element 6
print(b[1][2])

```

```python
print(c)

# getting element 6
print(c[1][1][0])
```

```python
# slicing
print(a[2:5])

#slicing with steps
print(a[2:5:2])

#first row of 2d array
print(b[0])

#first row and all cols of 2d array
print(b[0,:])

# only the third col
print(b[:,2])
```

```python
print(b)
# getting 5, 6, 9, 10
print(b[1:, 1:3])

# getting 0, 3, 8, 11
# first we get alternate rows
# first we get cols with gap 2
print(b[::2, ::3])

# getting 1, 3, 9, 11
print(b[::2, 1::2])

# getting 4, 7
print(b[1, ::3])
```

```python
print(c, "\n\n")

# the middle 2d array
print(c[1])

# second row of 1st 2d array : 3, 4, 5
print(c[0][1])
print(c[0, 1])

# getting 10, 13, 16
print(c[1, :, 1 ])

# getting matrix 22 23 25 26
print(c[2, 1:, 1:])

# ### getting 0, 2, 18, 20
print(c[::2, 0, ::2])
```

## Reshaping : reshape, transpose, ravel

```python
print(b)

#reshape
print(b.reshape(6,2))

# transpose
print(b.T) #or
print(np.transpose(b))

#ravel : converts n dimension array to 1d
print(b.ravel())
print(c.ravel())
```

## stacking : joining arrays (same shape required)

```python
# horizontal stacking
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12, 24).reshape(3,4)
print(a1, a2)
print(np.hstack((a1, a2, a1, a2)))
```

```python
# vertical stacking
print(np.vstack((a1,a2,a1,a2)))
```

## Splitting

```python
# horizontal splitting (we have to provide parts)
d = np.hstack((a1, a2, a1, a2))
print(d)
print(np.hsplit(d, 4))

# vertical splitting
#print(np.vsplit(b, 2))
```
